```python
# import modules
import os
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
plots_dir = os.path.join('plots/misc')
if not os.path.exists(plots_dir):
    os.system('mkdir -p {}'.format(plots_dir))
```

```python
# tanh annealing function
# this function ramps up from 0 to 1 in tanh(x) between the steps: fake_discriminator_warmup_steps and self.config.max_steps)
def anneal_tanh(step, warmup_steps, max_steps):
    return ((math.tanh((2.0 * ((step * 2.0 - ((max_steps + warmup_steps)))) / 
                              (max_steps - warmup_steps))) + 1.0) / 2.0)
```

```python
max_steps = 250000
warmup_steps = 10000
anneal_factor = np.array([anneal_tanh(s, warmup_steps, max_steps) for s in range(max_steps)])
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(max_steps), anneal_factor)
# plt.title('Heated Tanh Annealing || warmup steps: {} || max steps: {}'.format(warmup_steps, max_steps), fontsize=16)
plt.grid(color='r', linestyle='--', linewidth=2, alpha=0.2)
plt.savefig('{}/heated.tanh.annealing.warmup.steps.{}.max.steps.{}.png'.format(plots_dir, warmup_steps, max_steps))
```
